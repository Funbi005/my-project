```python
#data cleaning
import pandas as pd
data = pd.read_csv('boat_data.csv')
# Dropping duplicates values
data = data.drop_duplicates()
# Dropping null values
data = data.dropna()
# dropping The unlogical values
data= data[data["Year Built"]!=0]
# Creating two columns (Currency , Prices) from column (Price) to separate each of them 
data[['Currency','Prices']] = data['Price'].str.split(' ', expand=True)
# Seting the data type of prices as integar
data['Prices']= data['Prices'].astype('int')
# Creating a new column (Size) by multiplying the (Length and Width) Columns
data['Size'] = round(data['Width'].mul(data['Length']),2)
# Creating (Country) column by Separating the Country from (location) column 
data['Country'] = data['Location'].str.split('Â| Â', n=1, expand= True)[0]
# Setting the data type of (Year Bulit) as integar
data['Year Built']= data['Year Built'].astype('int')
# adding column Age which the age of the boat
data['Boat_age'] = 2022 - data['Year Built']
data.head(100)
```

```python
data.describe()
```

# pyplot visualization


```python

import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
```

```python
#To get the price of the most viewed boat

fig = plt.figure(figsize = (10, 5))
data.groupby("Prices")["Number of views last 7 days"].sum().sort_values(ascending =False).head(10).plot(kind="bar", fontsize=20, color= [ "purple", "purple", "purple", "purple", "purple", "purple", "purple", "purple", "purple", "purple" ])
ax = plt.gca()
ax.set_title('Top Viewed prices',fontsize= 20)
ax.set_xlabel('Prices',fontsize=20)
ax.set_ylabel("Views", fontsize=20)
ax.bar_label(ax.containers[0], label_type='edge',fontsize= 13)
plt.show()
```

```python
#To get the frequency of the prices viewed the most
_ , ax = plt.subplots()
data['Prices'].value_counts()
data['Prices'].value_counts().sort_values(ascending =False).head(10).plot(ax=ax, kind= 'bar', xlabel = 'Boat prices', ylabel= 'frequency')
plt.show()
```

```python
#To get the currency that has more view in the last 7 days

fig = plt.figure(figsize = (10, 5))
data.groupby("Currency")["Number of views last 7 days"].sum().sort_values(ascending =False).head(10).plot(kind="bar", fontsize=20, color= ["purple", "purple", "purple", "purple"])
ax = plt.gca()
ax.set_title('Top Viewed currency',fontsize= 20)
ax.set_xlabel('Currency',fontsize=20)
ax.set_ylabel("Views in the last 7 days", fontsize=20)
ax.bar_label(ax.containers[0], label_type='edge',fontsize= 13)
plt.show()
```

```python
#plot for boat type vs number of views last 7 days

fig = plt.figure(figsize = (10, 5))
#grouping and calculating aggregate value
data.groupby("Boat Type")["Number of views last 7 days"].sum().sort_values(ascending =False).head(10).plot(kind="bar", fontsize=20, color= [ "purple", "purple", "purple", "purple", "purple", "purple", "purple", "purple", "purple", "purple" ])
ax = plt.gca()
ax.set_title('Top Viewed boat type',fontsize= 20)
ax.set_xlabel('Boat type',fontsize=20)
ax.set_ylabel("Views of boat", fontsize=20)
ax.bar_label(ax.containers[0], label_type='edge',fontsize= 13)
plt.grid()
plt.show()
```

```python
#plot for the manufacturer that has the most view in the last 7 days

fig = plt.figure(figsize = (10, 5))
data.groupby("Manufacturer")["Number of views last 7 days"].sum().sort_values(ascending =False).head(10).plot(kind="bar", fontsize=20, color= [ "purple", "red", "purple", "red", "purple", "red", "purple", "red", "purple", "red" ])
ax = plt.gca()
ax.set_title('Top Manufacturers',fontsize= 20)
ax.set_xlabel('Manufacturers',fontsize=20)
ax.set_ylabel("views", fontsize=20)
ax.bar_label(ax.containers[0], label_type='edge',fontsize= 12)
plt.show()
```

```python
#plot for the boat material that has the most view in the last 7 days

fig = plt.figure(figsize = (10, 5))
data.groupby("Material")["Number of views last 7 days"].sum().sort_values(ascending =False).head(10).plot(kind="bar", fontsize=20, color= [ "purple", "red", "purple", "red", "purple", "red", "purple", "red", "purple", "red" ])
ax = plt.gca()
ax.set_title('Top viewed Materials',fontsize= 20)
ax.set_xlabel('Materials',fontsize=20)
ax.set_ylabel("views", fontsize=20)
ax.bar_label(ax.containers[0], label_type='edge',fontsize= 12)
plt.show()
```

```python
#plot for the boat size that has the most view in the last 7 days

fig = plt.figure(figsize = (10, 5))
data.groupby("Size")["Number of views last 7 days"].sum().sort_values(ascending =False).head(10).plot(kind="bar", fontsize=20, color= [ "purple", "red", "purple", "red", "purple", "red", "purple", "red", "purple", "red" ])
ax = plt.gca()
ax.set_title('Top Viewed sizes',fontsize= 20)
ax.set_xlabel('Sizes',fontsize=20)
ax.set_ylabel("Views", fontsize=20)
ax.bar_label(ax.containers[0], label_type='edge',fontsize= 13)
plt.show()
```
